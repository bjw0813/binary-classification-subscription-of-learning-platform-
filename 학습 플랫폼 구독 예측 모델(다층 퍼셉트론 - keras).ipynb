```python
from keras.optimizers import Adam
from keras.models import Sequential,Model
from keras import layers
from keras import Input
from keras import optimizers
from keras.layers import Dense
from keras.layers import concatenate
from keras.layers import Dropout
```

```python
model_DL = Sequential()
model_DL.add(Dense(128, activation = 'relu'))#뉴런 개수를 크게 하면 정확도는 증가할 수 있으나, 과적합이 발생할 수 있다.
model_DL.add(Dense(64, activation = 'relu'))
model_DL.add(Dropout(0.2))#드롭아웃을 쓰면 과적합 걱정을 줄일 수 있지만, 너무 드롭아웃 값을 크게 잡으면 성능이 떨어질 수 있다.
model_DL.add(Dense(32, activation = 'relu'))
model_DL.add(Dropout(0.2))
model_DL.add(Dense(16, activation = 'relu'))
model_DL.add(Dense(4, activation="relu"))
model_DL.add(Dense(1, activation = 'sigmoid'))
#이진 분류에는 보통 시그모이드를 쓴다. 다중 분류에는 소프트맥스를 쓴다. 일반적인 회귀에는 linear를 쓴다.
```

```python
model_DL.compile(loss="binary_crossentropy", optimizer="adam", metrics=["binary_accuracy"]) #optimizer=optimizers.SGD(learning_rate=0.005)
model_DL.fit(x_train,y_train,epochs=200,batch_size=32 ,verbose=1)
score = model_DL.evaluate(x_test, y_test)
```
